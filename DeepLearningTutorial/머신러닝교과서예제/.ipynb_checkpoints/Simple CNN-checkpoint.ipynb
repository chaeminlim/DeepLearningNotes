```python
import tensorflow as tf
import numpy as np
```

```python
mnist = tf.keras.datasets.mnist
(x_data, y_data), (x_test, y_test) = mnist.load_data()

print('{}'.format(x_data.shape))
print('{}'.format(x_test.shape))

x_train, y_train = (x_data[:50000, :]).reshape((50000, -1)), (y_data[:50000]).reshape((50000, -1))
x_valid, y_valid = (x_data[50000:, :]).reshape((10000, -1)), (y_data[50000:]).reshape((10000, -1))
x_test, y_test = (x_test).reshape((10000, -1)), (y_test).reshape((10000, -1))

print('{}'.format(x_train.shape))
print('{}'.format(x_valid.shape))
print('{}'.format(x_test.shape))
```

```python
mean_vals = np.mean(x_train, axis = 0)
std_val = np.std(x_train)

x_train_centered = (x_train - mean_vals) / std_val
x_valid_centered = (x_valid - mean_vals) / std_val
x_test_centered = (x_test - mean_vals) / std_val
```

```python
x_train_centered = x_train_centered.reshape((-1, 28, 28, 1))
x_valid_centered = x_valid_centered.reshape((-1, 28, 28, 1))
x_test_centered = x_test_centered.reshape((-1, 28, 28, 1))
```

```python
from tensorflow.keras.utils import to_categorical

y_train_onehot = to_categorical(y_train)
y_valid_onehot = to_categorical(y_valid)
y_test_onehot = to_categorical(y_test)
```

```python
print(y_train[0], y_train_onehot[0])
```

```python
print('훈련셋', x_train_centered.shape, y_train_onehot.shape)
print('검증셋', x_valid_centered.shape, y_valid_onehot.shape)
print('테스트셋', x_test_centered.shape, y_test_onehot.shape)
```

## 텐서프롤 tf.keras API로 cnn 구성

텐서플로에서 cnn을 구현하기 위해 tf.keras API로 cnn 네트웍을 구현하기 위해 tf.keras의 layers, models를 임포트 한다.

```python
from tensorflow.keras import layers, models
```

먼저 Seqenctial 모델을 만든다.

```python
MyModel = models.Sequential()
```

스트라이드를 설정하는 strides 매개 변수는 정수 또는 두 개로 이루어진 튜플로 지정한다.
https://www.tensorflow.org/api_docs/python/tf/keras/layers/Conv2D?version=stable

활성화 함수는 activation 매개 변수를 통해 지정한다.
kernel_initializer와 bias_initializer 매개 변수는 따로 지정하지 않았기 때문에 기본값으로 설정된다.

```python
MyModel.add(layers.Conv2D(
    32, 
    (5, 5), 
    padding='valid', 
    activation='relu', 
    input_shape=(28, 28, 1)
))
MyModel.add(layers.MaxPool2D((2, 2)))
MyModel.add(layers.BatchNormalization())

MyModel.add(layers.Conv2D(64,
                       (5, 5),
                       padding='valid',
                       activation='relu'))
MyModel.add(layers.MaxPool2D((2, 2)))
MyModel.add(layers.BatchNormalization())
MyModel.add(layers.Conv2D(64,
                       (5, 5),
                       padding='same',
                       activation='relu'))
MyModel.add(layers.MaxPool2D((2, 2)))
MyModel.add(layers.BatchNormalization())

MyModel.add(layers.Flatten())
MyModel.add(layers.Dense(256, activation='relu'))
MyModel.add(layers.Dropout(0.5))
MyModel.add(layers.Dense(10, activation='softmax'))
```

```python
MyModel.summary()
```

## CNN 모델의 훈련

이제 모델을 컴파일 한다. 다중 분류 작업이므로 손실 함수는 이전 장에서 사용했던 것처럼 categorical_crossentropy를 사용한다. 옵티마이저는 adam을 사용한다.
또 손실 점수와 더불어 정확도 값을 계산하기 위해 metrics 매개 변수에 acc를 추가한다.

```python
MyModel.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

이 구현에서는 tf.keras.optimizer.Adam 클래스를 사용하여 훈련한다. adam 옵티마이저는 볼록하지 않은 함수의 최적화와 머신 러닝 문제에 잘 맞는 경사 하강법 기반의 안정적인 옵티마이저이다. adam 옵티마이저에 영향을 미친 두개의 최적화 방븝은 RMSProp과 adaGrid이다.


```python
import time
from tensorflow.keras.callbacks import ModelCheckpoint, TensorBoard

callback_list = [ModelCheckpoint(filepath='cnn_checkpoint.h5', monitor='val_loss', save_best_only=True),
                TensorBoard(log_dir='logs/{}'.format(time.asctime()))]
```

```python
history = MyModel.fit(x_train_centered, y_train_onehot, batch_size = 64, epochs = 20,
                     validation_data=(x_valid_centered, y_valid_onehot))
```

```python
import matplotlib.pyplot as plt

epochs = np.arange(1, 21)
plt.plot(epochs, history.history['loss'])
plt.plot(epochs, history.history['val_loss'])
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```

```python
plt.plot(epochs, history.history['acc'])
plt.plot(epochs, history.history['val_acc'])
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.show()
```
